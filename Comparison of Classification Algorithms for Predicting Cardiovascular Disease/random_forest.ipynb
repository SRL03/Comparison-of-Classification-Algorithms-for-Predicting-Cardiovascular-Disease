```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df =pd.read_csv('cardio.csv',sep=';')
```

```python
df
```

```python
df["bmi"] = (df["weight"]/ (df["height"]/100)**2)
```

```python
df["age"] = df["age"]/365
```

```python
df["age"] = df["age"].astype("int")
```

```python
df["height"] = df["height"]/100
```

```python
df['pulse_pressure'] = df['ap_hi'] - df['ap_lo']
```

```python
df=df.drop(['id'],axis=1)
```

```python
df["gender"] = df["gender"] % 2
```

```python
df
```

```python
x = df[['age','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','bmi','pulse_pressure']]
```

```python
x
```

```python
y = df['cardio']
```

```python
y
```

```python
xtrain , xtest ,ytrain , ytest = train_test_split (x,y, test_size=.3,random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xtrain,ytrain)
```

```python
rfc.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score_rfc = cross_val_score(rfc,x,y,cv=10)
```

```python
cv_score_rfc
```

```python
cv_score_rfc.mean()
```

```python
pred_rfc = rfc.predict(xtest)
```

```python
pred_rfc
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred_rfc)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest,pred_rfc)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,pred_rfc))
```

```python
rfc_probs = rfc.predict_proba(xtest)
rfc_probs = rfc_probs[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
rfc_auc = roc_auc_score(ytest, rfc_probs)
```

```python
print('Random Forest: AUROC = %.3f' % (rfc_auc))
```

```python
rfc_fpr, rfc_tpr, _ = roc_curve(ytest, rfc_probs)
```

```python
plt.plot(rfc_fpr, rfc_tpr, marker='.', label='Random Forest (AUROC = %0.3f)' % rfc_auc)
# Title
plt.title('ROC Plot')
# Axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# Show legend
plt.legend() # 
# Show plot
plt.show()
```
