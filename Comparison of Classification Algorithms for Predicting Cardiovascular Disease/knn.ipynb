```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df =pd.read_csv('cardio.csv',sep=';')
```

```python
df
```

```python
df["bmi"] = (df["weight"]/ (df["height"]/100)**2)
```

```python
df["age"] = df["age"]/365
```

```python
df["age"] = df["age"].astype("int")
```

```python
df["height"] = df["height"]/100
```

```python
df['pulse_pressure'] = df['ap_hi'] - df['ap_lo']
```

```python
df=df.drop(['id'],axis=1)
```

```python
df["gender"] = df["gender"] % 2
```

```python
df
```

```python
x = df[['age','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','bmi','pulse_pressure']]
```

```python
x
```

```python
y = df['cardio']
```

```python
y
```

```python
xtrain , xtest ,ytrain , ytest = train_test_split (x,y, test_size=.3,random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
knn = KNeighborsClassifier(n_neighbors=265)
```

```python
knn.fit(xtrain,ytrain)
```

```python
knn.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score_knn = cross_val_score(knn,x,y,cv=10)
```

```python
cv_score_knn
```

```python
cv_score_knn.mean()
```

```python
pred_knn = knn.predict(xtest)
```

```python
pred_knn
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred_knn)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest,pred_knn)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,pred_knn))
```

```python
knn_probs = knn.predict_proba(xtest)
knn_probs = knn_probs[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
knn_auc = roc_auc_score(ytest, knn_probs)
```

```python
print('KNN: AUROC = %.3f' % (knn_auc))
```

```python
knn_fpr, knn_tpr, _ = roc_curve(ytest, knn_probs)
```

```python
plt.plot(knn_fpr, knn_tpr, marker='.', label='KNN(AUROC = %0.3f)' % knn_auc)
plt.title('ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend() 
plt.show()
```
