```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df =pd.read_csv('cardio.csv',sep=';')
```

```python
df
```

```python
df["bmi"] = (df["weight"]/ (df["height"]/100)**2)
```

```python
df["age"] = df["age"]/365
```

```python
df["age"] = df["age"].astype("int")
```

```python
df["height"] = df["height"]/100
```

```python
df['pulse_pressure'] = df['ap_hi'] - df['ap_lo']
```

```python
df=df.drop(['id'],axis=1)
```

```python
df["gender"] = df["gender"] % 2
```

```python
df
```

```python
x = df[['age','gender','height','weight','ap_hi','ap_lo','cholesterol','gluc','smoke','alco','active','bmi','pulse_pressure']]
```

```python
x
```

```python
y = df['cardio']
```

```python
y
```

```python
xtrain , xtest ,ytrain , ytest = train_test_split (x,y, test_size=.3,random_state=1)
```

```python
xtrain
```

```python
xtest
```

```python
tree =  DecisionTreeClassifier()
```

```python
tree.fit(xtrain,ytrain)
```

```python
tree.score(xtest,ytest)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score_tree = cross_val_score(tree,x,y,cv=10)
```

```python
cv_score_tree
```

```python
cv_score_tree.mean()
```

```python
pred_tree = tree.predict(xtest)
```

```python
pred_tree
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pred_tree)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(ytest,pred_tree)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,pred_tree))
```

```python
tree_probs = tree.predict_proba(xtest)
tree_probs = tree_probs[:, 1]
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
tree_auc = roc_auc_score(ytest, tree_probs)
```

```python
print('Desicion Tree: AUROC = %.3f' % (tree_auc))
```

```python
tree_fpr, tree_tpr, _ = roc_curve(ytest, tree_probs)
```

```python
plt.plot(tree_fpr, tree_tpr, marker='.', label='DT(AUROC = %0.3f)' % tree_auc)
plt.title('ROC Plot')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend() 
plt.show()
```
